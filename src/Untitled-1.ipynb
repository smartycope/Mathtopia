```r
f <- function(x, a, b){
    return(a*x^b)
}
```

```r
T <- function(a, b, A, B, C, D){
    rtn <- function(x){
        return(A * f(B*(x-C), a, b) + D)
    }
    return(rtn)
}
```

```r
matrix_data <- matrix(c(1, 2, 1, 1, 0, 0,
                        1, 2, 1, 1, 3, 4,
                        1, 2, 3, 4, 0, 0,
                        3, 4, 1, 1, 0, 0,
                        -1, 3, 2, -4, 5, -6), nrow = 5, byrow = TRUE)

apply(matrix_data, 1, function(row) {
  plot(T(row[1], row[2], row[3], row[4], row[5], row[6]))
})
```
